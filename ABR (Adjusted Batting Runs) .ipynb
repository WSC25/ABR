## 計算個人殘壘數

```python
sheet_name1 = 'AWAY-打席資訊'  
sheet_name2 = 'HOME-打席資訊'
```

```python
away = pd.read_excel(excel_file_path, sheet_name=sheet_name1)
home = pd.read_excel(excel_file_path, sheet_name=sheet_name2)
```

```python
cols = ['打者','壘包狀態','開始出局數','結束出局數'] 
```

```python
away_lob = away[cols]
home_lob = home[cols]
```

```python
LOB = pd.concat([home_lob,away_lob], ignore_index=False)
```

```python
LOB['出局數']=LOB['結束出局數']-LOB['開始出局數']
```

```python
LOB = LOB[LOB['出局數'] != 0]
LOB.loc[LOB['出局數'] == 2, '出局數'] = 1
```

```python
def classify_lob(row):
    if row['壘包狀態'] in [1,2,4]:
        return 1
    elif row['壘包狀態'] in [3,5,6]:
        return 2
    elif row['壘包狀態'] in [7]:
        return 3
    else:
        return 0


LOB['LOB'] = LOB.apply(classify_lob, axis=1)

LOB
```

```python
LOB= LOB.groupby(['打者']).agg({'LOB':'sum'})
LOB
```

## 球員個人成績整理

```python
sheet_name3 = 'AWAY-BOX-打擊'  
sheet_name4 = 'HOME-BOX-打擊'
```

```python
away = pd.read_excel(excel_file_path, sheet_name=sheet_name3)
home = pd.read_excel(excel_file_path, sheet_name=sheet_name4)
```

```python
merged_stat = pd.concat([away, home], ignore_index=True)
merged_stat.info()
```

```python
merged_stat = merged_stat.drop(['球場',"比賽編號","比賽日期","客場","主場","打序","背號",'滾地雙殺','雙殺','三殺','失誤上壘','故意四壞','犧短','犧飛'], axis = 1)
merged_stat
```

```python
agg_dict = {
    "打席": "sum",
    "打數": "sum",
    "得分": "sum",
    "安打": "sum",
    "打點": "sum",
    "二安": "sum",
    "三安": "sum",
    "全壘打": "sum",
    "四壞": "sum",
    "觸身": "sum",
    "三振": "sum",
    "盜壘": "sum",
    "盜壘失敗": "sum"
}
batter_total = merged_stat.groupby("名稱").agg(agg_dict).reset_index()
batter_total
```

```python
batter_total['一安'] = batter_total['安打']-batter_total['二安']-batter_total['三安']-batter_total['全壘打']
batter_total['TB'] = batter_total['一安']+batter_total['二安']*2+batter_total['三安']*3+batter_total['全壘打']*4
```

## 殘壘數和個人成績資料合併

```python
batter_total = batter_total.rename(columns={'名稱':'打者'})
```

```python
df = pd.merge(batter_total, LOB, on=['打者'])
```

## 係數整理+計算ABR(來自計算WAR時的RE24 四捨五入至小數點第三位)

## 原公式:ABR = (0.47)x1B + (0.78)x2B + (1.09)x3B + (1.40)xHR + (0.33)x(BB+HBP) - (0.25)x(AB-H) - (0.50)x(H+BB+HBP-LOB-R-CS)

## 2023 RE整理
### 一安 0.489
### 二安 0.804
### 三安 1.072
### HR 1.712
### BB 0.354
### HBP 0.373
#### 係數"0.25"和"0.5" 尚未做變動 需要另外處理 EX:0.25由來可能是計算打擊率和得分的關係得來 

```python
df['ABR(2023RE)']=0.489*df['一安']+0.804*df['二安']+1.072*df['三安']+1.712*df['全壘打']+0.354*df['四壞']+0.373*df['觸身']-0.25*(df['打數']-df['安打'])-0.5*(df['安打']+df['四壞']+df['觸身']-df['LOB']-df['得分']-df['盜壘失敗'])
```
